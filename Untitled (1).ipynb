```python
import pandas as pd
import re
```

```python
list_of_question=[]
list_of_option1=[]
list_of_option2=[]
list_of_option3=[]
list_of_option4=[]
fhand=open('Sample Question File_q.txt')
for line in fhand:
    line = line.rstrip()
    if not line.find('@@@') >=0 and not line.find('###0')>=0:
        if line.find('(a)')>=0:
            line=line.lstrip('(a)')
            list_of_option1.append(line)
        elif line.find('(b)')>=0:
            line=line.lstrip('(b)')
            list_of_option2.append(line)
        elif line.find('(c)')>=0:
            line=line.lstrip('(c)')
            list_of_option3.append(line)
        elif line.find('(d)')>=0:
            line=line.lstrip('(d)')
            list_of_option4.append(line)
        else:
            if line[0].isdigit():
                line=line[3:]
                list_of_question.append(line)
        
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame([list_of_question,list_of_option1,list_of_option2,list_of_option3,list_of_option4],index=['question no','option 1','option2','option3','option4'])
```

```python
df1=np.transpose(df1)
```

```python

ans=[]
i=1
fhand = open('Sample Solution File_s.txt')
for line in fhand:
    if not line.startswith('###'):
        if line[0].isdigit():
            line=line[3:10]
            ans.append(line)
```

```python
ans_key=[]
for w in ans:
    w=w.strip('\n')
    ans_key.append(w)
```

```python
ans_key=np.transpose(ans_key)
```

```python
df1['Answer Key']=ans_key
```
